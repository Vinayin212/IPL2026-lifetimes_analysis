```python
df['batting_efficiency'] = df['Runs_Scored'] / df['Batting_Strike_Rate']
batting_efficiency = df.sort_values(by='batting_efficiency', ascending=False)[['Player_Name', 'batting_efficiency']].head(10)
print(batting_efficiency)
sns.barplot(x='batting_efficiency', y='Player_Name', data=batting_efficiency, palette='viridis')    
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')    
plt.show()
```

```python
top_players= df.groupby('Player_Name')['Runs_Scored'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_players.values, y=top_players.index, palette='viridis')
plt.xlabel('Total Runs Scored')
plt.title('Top 10 Run Scorers in IPL 2026')
```

```python
sns.histplot(df['Runs_Scored'], bins=20, kde=True)
plt.title('Distribution of Runs Scored')    
```

```python
sns.scatterplot(x='Runs_Scored', y='Batting_Strike_Rate', data=df)
plt.title('Runs Scored vs Batting Strike Rate')
```

```python
df['All_Rounder_Score'] = (df['Runs_Scored'] / df['Batting_Average']) + (df['Wickets_Taken'] / df['Bowling_Average'])
top_all_rounders = df.sort_values(by='All_Rounder_Score', ascending=False)[['Player_Name', 'All_Rounder_Score']].head(10)
print(top_all_rounders)
```

```python
yearly_runs = df.groupby('Year')['Runs_Scored'].sum()
plt.figure(figsize=(10,6))  
sns.lineplot(x=yearly_runs.index, y=yearly_runs.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Total Runs Scored')
plt.title('Yearly Runs Scored in IPL 2026')
```

```python
df['Consistency_Score'] = (df['Runs_Scored'] / df['Batting_Average']) * (df['Wickets_Taken'] / df['Bowling_Average'])
top_consistent_players = df.sort_values(by='Consistency_Score', ascending=False)[['Player_Name', 'Consistency_Score']].head(10)
print(top_consistent_players)
sns.histplot(df['Consistency_Score'], bins=20, kde=True)
```

```python
s1= print("Total players:", df['Player_Name'].sum())
s2= print("Total runs:", df['Runs_Scored'].sum())
s3= print('total wickets:', df['Wickets_Taken'].sum())
sns.barplot(x=['Total Players', 'Total Runs', 'Total Wickets'], y=[s1, s2, s3], palette='Set2')
plt.title('Summary of IPL 2026 Data')
plt.ylabel('Count')
plt.show()
```

```python
ipl_summary = {
    'Total Players': df['Player_Name'].nunique(),
    'Total Runs': df['Runs_Scored'].sum(),
    'Total Wickets': df['Wickets_Taken'].sum(),

}
sns.barplot(y=list(ipl_summary.keys()), x=list(ipl_summary.values()), palette='Set3')
plt.title('IPL 2026 Summary Statistics')
    
```
